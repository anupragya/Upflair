# Object-Oriented Programming (OOP) 
 programming paradigm based on the concept of "objects", which can contain data and code. Data is stored in fields (attributes or properties), and code is stored in methods(functions associated with the object)

```python
# Class and object
class Student:
    name = "saksham"  # class variable
    age = 20          # class variable
    print(f"My name is {name} and my age is {age}")
obj_student = Student()
print(obj_student.name)  # Output: saksham
print(obj_student.age)   # Output: 20
```

```python
#Methods and Self Keyword
class Student:
    name = "nanus"
    age = 20
def info(self):  # Method
    print("This is my first method")
    print(f"My name is {self.name} and my age is {self.age}")
ob = Student()
ob.info()
# Explanation:self refers to the current instance of the class
```
