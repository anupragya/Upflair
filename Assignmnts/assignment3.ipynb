```python
#Q1: Wite a program that takes a list of numbers and prints all prime numbers in the list using loops. Do not use any built-in prime-checking functions. 

num_list = [1, 2,3,4,5,6,7,8,9]


prime_numbers = []


for num in num_list:
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            prime_numbers.append(num)

print("Prime numbers found:", prime_numbers)


```

```python
#Q10:  Write a program that reads numbers from a file and calculates: Total, Average, Handle the case if the file does not exist.
try:
    file= open("stu.txt", 'r')
    stu = [float(num) for num in file.read().split()]

    total = sum(stu)
    average = total / len(stu)

    print("Total =", total)
    print("Average =", average)

    file.close()
except FileNotFoundError:
    print("damn")
except ZeroDivisionError as z:
    print(z)
    
    
```
