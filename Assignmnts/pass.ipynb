```python
password= input("enter your password: ")
for i in password:
    ch= password[i]
    if ((ch>='a' and ch<='z') or (ch>='A' and ch<='Z')) and (ch>='0' and ch<=9) and not password.isalnum():
        print("password is strong")
    else:
        print("weak password")
```

```python
def counting(s):
    vo=0
    con=0
    cha=0
    dig=0
    for i in range(0, len(s)): 
        ch = s[i]
        if (ch>='a' and ch<='z') or (ch>='A' and ch<='Z'):
            ch = ch.lower()
            if ch in "aeiouAEIOU":
                vo+=1
            else:
                con+=1
        elif (ch>='0' and ch<='9'):
            dig+=1
        else:
            cha+=1
    print("Vowels:", vo)
    print("Consonant:", con) 
    print("Digit:", dig) 
    print("Special Character:", cha) 
```
