```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
plt.scatter(df['tip'], df['total_bill'])
plt.xlabel("tips")
plt.ylabel("total bill")
plt.show()
```

```python
sns.scatterplot(x='tip', y='total_bill', data=df)

plt.show()
```

```python
plt.hist(df['total_bill'], color="pink", width=10)
plt.show()
```

```python
sns.histplot(df['total_bill'])
plt.show()
```

```python
sns.kdeplot(df['total_bill'])
plt.show()
```

```python
sns.relplot(x='tip', y='total_bill', data=df)
plt.show()
```

```python
sns.relplot(x='tip', y='total_bill', data=df, hue='smoker')
plt.show()
```

```python
sns.relplot(x='tip', y='total_bill', data=df, hue='time')
plt.show()
```

```python
sns.relplot(x='tip', y='total_bill', data=df, hue='time', style='smoker')
plt.show()
```

```python
sns.relplot(x='tip', y='total_bill', data=df, hue='time', style='smoker', col='day')
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=df, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=df, kind='box')
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=df, kind='boxen')
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=df, kind='violin')
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=df, kind='strip')
plt.show()
```

```python
sns.barplot(x='total_bill', y='day', data=df)
```

```python
sns.catplot(x='smoker', y='sex', data=df, kind='violin')
plt.show()
```
