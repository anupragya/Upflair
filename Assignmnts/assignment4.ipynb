```python
# Create a 5 × 5 NumPy matrix containing numbers from 1 to 25 and print: ○ The first row ○ The last column ○ The diagonal elements 
import numpy as np

# 1. Create a 5x5 matrix containing numbers from 1 to 25
matrix = np.arange(1, 26).reshape(5, 5)
print("Original 5x5 Matrix:")
print(matrix)


# 2. Print the first row (Index 0)
first_row = matrix[0, :]
print("First Row:")
print(first_row)


# 3. Print the last column (Index -1)
last_column = matrix[:, -1]
print("Last Column:")
print(last_column)


# 4. Print the diagonal elements
diagonal_elements = np.diag(matrix)
print("Diagonal Elements:")
print(diagonal_elements)
```

```python
#Create a 6 × 6 matrix and print: ○ Main diagonal ○ Upper triangular part ○ Lower triangular part 
matrix=np.random.randint(1,26, 25).reshape(5,5)
print(matrix)
print("main diagonal:", np.diag(matrix))
print("upper triangular part:\n", np.triu(matrix))
print("lower triangular part:\n", np.tril(matrix))
```

```python
#Create a 5 × 5 matrix and find the position (row and column index) of the largest element. 
matrix=np.random.randint(1,26, 25).reshape(5,5)
print(matrix)
print("maximum value:", np.max(matrix))
fi=np.argmax(matrix)
row_idx, col_idx = np.unravel_index(fi, matrix.shape)
print(f"Flat Index position: {fi}")
print(f"2D Coordinates -> Row Index: {row_idx}, Column Index: {col_idx}")
```

```python
#Create a dataset of students' marks using a NumPy array and find: ● Highest marks ● Lowest marks ● Average marks ● Students scoring above average 

np.random.seed(7)
marks = np.random.randint(40, 100, size=15)

print("Students' Marks Dataset:")
print(marks)
print("-" * 45)

# 2. Find Highest, Lowest, and Average marks
highest_marks = np.max(marks)
lowest_marks = np.min(marks)
average_marks = np.mean(marks)

print(f"Highest Marks : {highest_marks}")
print(f"Lowest Marks  : {lowest_marks}")
print(f"Average Marks : {average_marks:.2f}")
print("-" * 45)

# 3. Find students scoring above average using boolean indexing
above_average_mask = marks > average_marks
students_above_avg = marks[above_average_mask]

print(f"Number of students scoring above average: {len(students_above_avg)}")
print("Marks of students scoring above average:")
print(students_above_avg)
```
