```python
#Question 1
numbers = []


print("Enter 10 numbers:")
for i in range(10):
    num = float(input(f"Number {i+1}: "))
    numbers.append(num)


numbers.sort()

smallest = numbers[0]          # First element
largest = numbers[-1]          # Last element
second_largest = numbers[-2]   # Second to last element


print(f"Smallest number: {smallest}")
print(f"Largest number: {largest}")
print(f"Second largest number: {second_largest}")
```

```python
#question 2: Count vowels, consonants, digits, and special characters in a string.  
def counting(s):
    vo=0
    con=0
    cha=0
    dig=0
    for i in range(0, len(s)): 
        ch = s[i]
        if (ch>='a' and ch<='z') or (ch>='A' and ch<='Z'):
            ch = ch.lower()
            if ch in "aeiouAEIOU":
                vo+=1
            else:
                con+=1
        elif (ch>='0' and ch<='9'):
            dig+=1
        else:
            cha+=1
    print("Vowels:", vo)
    print("Consonant:", con) 
    print("Digit:", dig) 
    print("Special Character:", cha) 

counting("hello112@") 
```

```python
#question 3
def is_armstrong(num):
    num_s = str(num)
    order = len(num_s)
    
    return num == sum(int(digit) ** order for digit in num_s)

print(is_armstrong(370))  
```

```python
#Question 6 :Count the frequency of each word in a sentence using a dictionary.  
def count_wrd(sen):
    sen = sen.lower()
    wrds= sen.split()
    wrd_freq={}
    for wrd in wrds:
        wrd_freq[wrd]= wrd_freq.get(wrd, 0)+1
    return len(wrd_freq)
count_wrd("hi my name is oshi")
```

```python
#Question 7: Display students scoring above the average marks from a dictionary. 
students={"anu": 85, "pratibha": 89, "sam": 98, "yugansh":87, "badha":100}
avg= sum(students.values())/len(students)
print(avg)
for name, score in students.items():
    if score >avg:
       print("student with above avg marks:", name)
```

```python
#Question 8: Create a menu-driven product management system using a dictionary.  
products = {}

while True:
    print("\n1. Add Product")
    print("2. View Products")
    print("3. Update Product")
    print("4. Delete Product")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        pid = input("Enter Product ID: ")
        pname = input("Enter Product Name: ")
        products[pid] = pname
        print("Product Added!")

    elif choice == "2":
        print("Products:", products)

    elif choice == "3":
        pid = input("Enter Product ID to update: ")
        if pid in products:
            products[pid] = input("Enter new Product Name: ")
            print("Product Updated!")
        else:
            print("Product not found!")

    elif choice == "4":
        pid = input("Enter Product ID to delete: ")
        if pid in products:
            del products[pid]
            print("Product Deleted!")
        else:
            print("Product not found!")

    elif choice == "5":
        print("Exiting...")
        break

    else:
        print("Invalid Choice!")
```

```python
#Question 9: Check whether a triangle is Equilateral, Isosceles, or Scalene.
fs= int(input("enter first side of triangle: "))
ss= int(input("enter second side of triangle: "))
ts= int(input("enter third side of triangle: "))

if fs==ss and ss==ts:
    print("equilateral")
elif fs!=ss and ss!=ts and fs!=ts:
    print("scalene")
else:
    print("isosceles")
```

```python
#Question 10:  Create a Student Management System using loops, tuples, dictionaries, 
# and conditions. 
students = {}

while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        roll = input("Enter Roll Number: ")
        name = input("Enter Name: ")
        marks = int(input("Enter Marks: "))
        students[roll] = (name, marks)      # Tuple
        print("Student Added Successfully!")

    elif choice == "2":
        if students:
            for roll, data in students.items():   # Loop
                print("Roll No:", roll,
                      "Name:", data[0],
                      "Marks:", data[1])
        else:
            print("No student records found.")

    elif choice == "3":
        roll = input("Enter Roll Number to Search: ")
        if roll in students:      # Condition
            print("Student Found:", students[roll])
        else:
            print("Student Not Found!")

    elif choice == "4":
        roll = input("Enter Roll Number to Delete: ")
        if roll in students:
            del students[roll]
            print("Student Deleted Successfully!")
        else:
            print("Student Not Found!")

    elif choice == "5":
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice!")
```
