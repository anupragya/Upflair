```python
import spacy
nlp = spacy.load("en_core_web_sm") ##python -m spacy download en_core_web_sm
text="   hello12 let's study ##NLP"
def text_cleaing(text):
    clean_data= "".join(char for char in text if char.isalpha() or char.isspace())
    result=nlp(clean_data)
    return result
text_cleaing(text=text)
```
