# *Exception Handling* 
allows a program to handle unexpected errors during execution in a controlled way, instead of crashing abruptly. It enables programs to detect errors, manage them properly and continue execution wherever possible.

```python
#types of error
# value error
try:
    a= int(input())
    b= int(input())
    print(a * b)
except ValueError as e:
    print("Invalid input:", e)
```

```python
#zerodivision error
try:
    a = int(input("Enter 1st number: "))
    b  = int(input("Enter 2nd number: "))
    print(a / b)
except ZeroDivisionError as e:
    print(e)
```

```python
#catching all exceptions
try:
   a = int(input("Enter 1st number: "))
   b = int(input("Enter 2nd number: "))
   print(a / b)
except Exception as e:
   print(e)
```

```python
#Using else and finally
try:
   a = int(input("Enter 1st number: "))
   b = int(input("Enter 2nd number: "))
   print(a / b)
except Exception as e:
   print(e)
else:
   print("No error")
   print(a * b)
finally:
   print("I will always run!")    


```

```python
#Multiple except Blocks
try:
   num1 = int(input("Enter a number: "))
   num2 = int(input("Enter another number: "))
   result = num1 / num2
except ZeroDivisionError as e:
   print(e)
except ValueError as e:
   print(e)
except Exception as e:
   print(e)
else:
   print("Final result is:", result)
finally:
   print("Thanks for using our service!")
```
