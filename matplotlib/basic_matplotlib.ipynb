```python
xpoint= np.array([1,2,3,4])
ypoint= np.array([2,4,6,8])
colour=np.array(["black", "blue", "pink", "purple"])
size =np.array([90,70,30,40])
plt.xlabel("xpoint")
plt.ylabel("ypoint")
plt.title("matplotlib is amazing")
plt.scatter(xpoint, ypoint, c=colour, s= size, alpha=0.5)
plt.show()
```
