# *Conditional Statement*
It is programming construct that executes a specific block of code when a given condition is satisfied. 

```python
print([i for i in range(3)])



```

```python
age = int(input("Enter your age👉"))
if(age>=18):
    print("you are eligible to vote")
else:
    print("not eligible")
```

```python
num= int(input("enter a number: "))
if num%2==0:
    print("It is even")
else:
    print("It is odd")
```

```python
amount= int(input("enter purchased amount:"))
if amount>5000:
    print("you get 20% discount")
elif amount>2000:
    print("you get 10% discount")
else:
    print("you get no discount")
```

```python
a=int(input("enter first number: "))
b=int(input("enter second number: "))   
if a>b:
    print("a is greater than b")
elif a<b:
    print("b is greater than a")
```

```python
year= int(input("enter year: "))
if (year%4==0 and year%100!=0) or (year%400==0):
    print("leap year")
else:    
    print("not a leap year")   
```

```python
a = int(input("enter number: "))
if(a%5==0 and a%11==0):
    print("it is divisible by 5 and 11")
else:
    print("it's not")
```
