# Inheritance
**Inheritance allows a class to acquire properties and methods of another class.**

```python
# TYPES OF INHERITANCE: SINGLE, MULTIPLE, MULTI LEVEL, HYBRID, HERARCHIAL 
```

 # *Single Inheritance*

```python
#syntax
class Company:
    pass
class Developer(Company):
    pass
```

```python
class Company:
    cpny_server_ip= "10.20.30"
    cpny_domain=".com"
class Developer(Company):
    dvlpr_name= "oshi"
    dvlpr_age= 21
    dvlpr_process="data science"
dev=Developer()
print(dev.cpny_server_ip)
print(dev.cpny_domain)
```

# *Multilevel Inheritance*

```python
class Company:
    cpny_server_ip= "10.20.30"
    cpny_domain=".com"


class Developer(Company):
    dvlpr_name= "anu"
    dvlpr_age= 21
    dvlpr_process="data science"

class Intern(Developer):
    intern_name= "oshi"
    intern_age= 21
    intern_process="data science"


dev=Intern()
print("company info:")
print(dev.cpny_server_ip)
print(dev.cpny_domain)
print("developer info:")
print(dev.dvlpr_age)
print(dev.dvlpr_name)
print(dev.dvlpr_process)
print("intern info:")
print(dev.intern_name)
print(dev.intern_age)
```

# *Multiple Inheritance*

```python
class Company:
    cpny_server_ip= "10.20.30"
    cpny_domain=".com"


class Developer():
    dvlpr_name= "anu"
    dvlpr_age= 21
    dvlpr_process="data science"

class Intern(Company, Developer):
    intern_name= "oshi"
    intern_age= 21
    intern_process="data science"

dev=Intern()
print("company info:")
print(dev.cpny_server_ip)
print(dev.cpny_domain)
print("developer info:")
print(dev.dvlpr_age)
print(dev.dvlpr_name)
print(dev.dvlpr_process)
print("intern info:")
print(dev.intern_name)
print(dev.intern_age)
```

# *Hybrid Inheritance*

#

```python
class Company:
    cpny_server_ip= "10.20.30"
    cpny_domain=".com"


class Developer(Company):
    dvlpr_name= "anu"
    dvlpr_age= 21
    dvlpr_process="data science"

class Intern(Company):
    intern_name= "oshi"
    intern_age= 21
    intern_process="data science"

class System(Company, Developer, Intern):
    pass


dev=System()
print("company info:")
print(dev.cpny_server_ip)
print(dev.cpny_domain)
print("developer info:")
print(dev.dvlpr_age)
print(dev.dvlpr_name)
print(dev.dvlpr_process)
print("intern info:")
print(dev.intern_name)
print(dev.intern_age)
```

# Polymorphism
Polymorphism in Object-Oriented Programming (OOP) is the ability for different objects to respond to the same method call in their own unique way. The word comes from Greek, meaning "many forms". It allows you to treat objects of various types as if they were a common type, making code highly flexible and reusable

```python
#_a=10
class A:
    _a=10
class B(A):
    pass
l=B()
l._a
```
