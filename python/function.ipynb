```python
#lambda function: A lambda function in Python is a small, anonymous function that is defined in a single line without a name. You use them for quick, short-lived operations where writing a full standard function using the def keyword would be unnecessary overhead
add = lambda a,b:a+b
print(add(1,5))
```

```python
def max_of_three(a, b, c):
    if (a>b) and (a>c):
        print(a)
    elif(b>a) and (b>c):
        print(b)
    else:
        print(c)
max_of_three(1,5,3)       
```

```python
def count_vowels(s):
    count=0
    for i in s:
        if i in "aeiouAEIOU":
            #print(i, end="")
            count+=1
    print(count)
count_vowels("Hello, World!")
```

```python
def factorial(n):
    if (n==0):
        return 1
    elif(n==1):
        return 1
    else:
        return n*factorial(n-1) #5*4*3*2*1
        
factorial(5)

```

```python
def apply_twice(func, x):
    return(func(func(x)))
def double(x):
    return x*2

apply_twice(double, 3)
```
