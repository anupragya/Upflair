# File Handling
File handling lets you read and write files on your computer.


```python
file= open("new.txt", 'x')
file.write("hey my name is oshi")
file.close()
```

```python
file = open("new.txt", "w")
file.write("hello") #overwrites in the file
file.close()
```

```python
file = open("new.txt", "a")
file.write("\nmy name is oshi") #writes in the file without erasing previous text
file.close()
```

```python
file = open("new.txt", 'r')
print(file.read())
file.close()
```
